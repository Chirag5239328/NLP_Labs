```python
import gensim.downloader as api
from gensim.models import Word2Vec, KeyedVectors
```

```python
wv=api.load("word2vec-google-news-300")
print(wv.most_similar("dollar"))
```

```python
words = ["computer", "science", "machine", "intelligence", "data"]

# Find and print similar words for each word in the list
for word in words:
    print(f"Most similar words to '{word}':")
    similar_words = wv.most_similar(word)
    for similar_word, similarity in similar_words:
        print(f"  {similar_word}: {similarity:.4f}")
    print()
```

```python
import pandas as pd
```

```python
df=pd.read_csv("IMDB Dataset.csv")
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
df["sentiment"]=l.fit_transform(df["sentiment"])
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
def preprocess_text(text):
    tokens = word_tokenize(text)
    filtered_tokens = [word for word in tokens if word and word not in stop_words]
    return filtered_tokens

df['review'] = df['review'].apply(preprocess_text)
    
```

```python
import numpy as np
def get_embedding(words):
    embedding = [wv[word] for word in words if word in wv]
    return np.mean(embedding, axis=0)

embed=df['review'].apply(get_embedding)  
```

```python
df1=pd.DataFrame(embed.tolist())
```

```python
df1
```

```python
finaldf=pd.concat([df,df1],axis=1)
finaldf
```

```python
sentences = df['review'].tolist()
sentences

skipgram = Word2Vec(
    sentences=sentences, # list of list of tokens
    sg=1,
    vector_size=50,
    window=5,
    min_count=1,
)
skipgram.save("skipgram.model")
```

```python
import numpy as np
sg = Word2Vec.load("skipgram.model")

# Access word vectors
word_vectors = sg.wv
def get_embedding(words):
    embedding = [word_vectors[word] for word in words if word in word_vectors]
    return np.mean(embedding, axis=0)

embedding=df['review'].apply(get_embedding)

data=pd.DataFrame(embedding.tolist())
final_df_skip=pd.concat([df,data],axis=1)
final_df_skip
```

```python
cbow = Word2Vec(
    sentences=sentences, # list of list of tokens
    sg=0,
    vector_size=50,
    window=5,
    min_count=1,
)
cbow.save("cbow.model")
model = Word2Vec.load("cbow.model")

# Access word vectors
word_vectors = model.wv
def get_embedding(words):
    embedding = [word_vectors[word] for word in words if word in word_vectors]
    return np.mean(embedding, axis=0)

embedding=df['review'].apply(get_embedding)

data=pd.DataFrame(embedding.tolist())
final_df_cbow=pd.concat([df,data],axis=1)
final_df_cbow
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(finaldf.iloc[:,2:], finaldf['sentiment'], test_size=0.2)
X_train_sk, X_test_sk, y_train_sk, y_test_sk = train_test_split(final_df_skip.iloc[:,2:], final_df_skip['sentiment'], test_size=0.2)
X_train_cb, X_test_cb, y_train_cb, y_test_cb = train_test_split(final_df_cbow.iloc[:,2:], final_df_cbow['sentiment'], test_size=0.2)


```

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier()
clf.fit(X_train,y_train)
clf_sk=RandomForestClassifier()
clf_sk.fit(X_train_sk,y_train_sk)
clf_cb=RandomForestClassifier()
clf_cb.fit(X_train_cb,y_train_cb)

```

```python
from sklearn.metrics import accuracy_score,f1_score
y_pred=clf.predict(X_test)
x=f1_score(y_test,y_pred)
y=accuracy_score(y_test,y_pred)
print("Accuracy:",x)
print("F1 Score:",y)

y_pred_sk=clf_sk.predict(X_test_sk)
x=f1_score(y_test_sk,y_pred_sk)
y=accuracy_score(y_test_sk,y_pred_sk)
print("Accuracy:",x)
print("F1 Score:",y)


y_pred_cb=clf_cb.predict(X_test_cb)
x=f1_score(y_test_cb,y_pred_cb)
y=accuracy_score(y_test_cb,y_pred_cb)
print("Accuracy:",x)
print("F1 Score:",y)
```
